```python
# POSITION BASED INDEXING
df8 = pandas.read_csv("http://pythonhow.com/supermarkets.csv")
df8.set_index("ID")
```

```python
# POSITION BASED INDEXING continued

# df8.iloc[:3]

df8.iloc[3:7,3:]
```

```python
# we can use ix and give the index and the column name directly
df8.ix[2,"Name"]
```

```python
# deleting column : 1 here lets drop function know that we are deleting column
df8.drop("City",1)
```

```python
# row 3 has been deleted, 0 shows that we are deleting rows
df8.drop(3,0)
```

```python
# we can also delete rows usind index.
df8.drop(df8.index[0:3],0)
```

```python
df8.drop(df8.columns[0:3],1)
```

```python
# ADDING COLUMN OR ROWS TO THE DATAFRAME
# adding new columns
# df.shape[0] give the number of rows
df8["Continent"] = df8.shape[0]*["North America"]
df8
```
